```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/roxoho/Downloads/Data-pv/csv-data/DATASET_V4.csv')
df.head()
```

```python
df.info()
```

```python
corr = df.corr()[['Avg Power']].sort_values(by='Avg Power', ascending=False)

plt.figure(figsize = (8,10))
sns.heatmap(corr,annot = True,cmap = 'coolwarm',linewidth = 0.5)
plt.show()
```

```python
df_withoutdust = df.drop(columns=['PM2.5', 'PM10'])
corrwithoutdust = df_withoutdust.corr()[['Avg Power']].sort_values(by='Avg Power', ascending=False)
plt.figure(figsize=(8, 10))
sns.heatmap(corrwithoutdust, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation with Avg Power (Excluding Dust Components)')
plt.show()
```

```python
X_with = df.drop(columns = ['Avg Power'])
y = df['Avg Power']
```

```python
X_without = df_withoutdust.drop(columns = ['Avg Power'])
```

```python
from sklearn.model_selection import train_test_split

# Split data with dust components
X_train_with_dust, X_test_with_dust, y_train, y_test = train_test_split(X_with, y, test_size=0.2, random_state=42)

# Split data without dust components
X_train_without_dust, X_test_without_dust, y_train, y_test = train_test_split(X_without, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error
```

```python
# Initialize the Random Forest Regressor
model1 = RandomForestRegressor(random_state=42)

# Train the model
model1.fit(X_train_with_dust, y_train)
```

```python
# Make predictions
y_pred_withdust = model1.predict(X_test_with_dust)
```

```python
mse1 = mean_squared_error(y_test,y_pred_withdust)
print("Mean Squared Error:", mse1)
```

```python
import numpy as np

# Sample 500 points randomly for visualization
sample_indices = np.random.choice(len(y_test), size=500, replace=False)
y_test_sampled = y_test.iloc[sample_indices]
y_pred_sampled = y_pred_withdust[sample_indices]

plt.figure(figsize=(10, 6))
plt.scatter(y_test_sampled, y_pred_sampled, alpha=0.5)
plt.plot([min(y_test_sampled), max(y_test_sampled)], [min(y_test_sampled), max(y_test_sampled)], 'r--', lw=2)
plt.xlabel('Actual Avg Power')
plt.ylabel('Predicted Avg Power')
plt.title('Actual vs Predicted Avg Power (Sampled, with Dust Components)')
plt.show()
```
